```python
# Calcular ECR
def calc_ecrs(dir_excel_5p):
    excel_5p = dir_excel_5p
    #sort by ix
    excel_5p = excel_5p.sort_values(by='IX',axis = 0, ascending=True)

    exp_pl = excel_5p['exp_MS_PL'].tolist()
    exp_dock6 = excel_5p['exp_MS_dock6'].tolist()
    exp_vina = excel_5p['exp_MS_vina'].tolist()
    exp_ad = excel_5p['exp_MS_AD'].tolist()

    # Combinaciones de 2 programas
    excel_5p["ecr_ad_dock6"] = calc_ecr2(exp_ad, exp_dock6, dir_excel_5p)
    excel_5p["ecr_ad_pl"] = calc_ecr2(exp_ad, exp_pl, dir_excel_5p)
    excel_5p["ecr_ad_vina"] = calc_ecr2(exp_ad, exp_vina, dir_excel_5p)
    excel_5p["ecr_dock6_pl"] = calc_ecr2(exp_dock6, exp_pl, dir_excel_5p)
    excel_5p["ecr_dock6_vina"] = calc_ecr2(exp_dock6, exp_vina, dir_excel_5p)
    excel_5p["ecr_pl_vina"] = calc_ecr2(exp_pl, exp_vina, dir_excel_5p)

    # coincidentsinaciones de 3 programas
    excel_5p["ecr_ad_dock6_pl"] = calc_ecr3(exp_ad, exp_dock6, exp_pl, dir_excel_5p)
    excel_5p["ecr_ad_dock6_vina"] = calc_ecr3(exp_ad, exp_dock6, exp_vina, dir_excel_5p)
    excel_5p["ecr_ad_pl_vina"] = calc_ecr3(exp_ad, exp_pl, exp_vina, dir_excel_5p)
    excel_5p["ecr_dock6_pl_vina"] = calc_ecr3(exp_dock6, exp_pl, exp_vina, dir_excel_5p)

    # coincidentsinaciones de 4 programas
    excel_5p["ecr_ad_dock6_pl_vina"] = calc_ecr4(exp_ad, exp_dock6, exp_pl, exp_vina, dir_excel_5p)

    return excel_5p 

```

```python
def select_rmsds(df):
    rmsd_columns = [col for col in df.columns if "RMSD" in col]
    # Create a list of lists for the result, initializing each sublist
    final = [[] for _ in range(len(rmsd_columns)-1)]

    for i in range(len(rmsd_columns)-1):
        combinations = rmsd_columns[i].split("_")
        # Check if the third element is "NEW"
        if combinations[2] == "NEW":
            # Append directly to the final list
            final[i].append(combinations[1] + "_" + combinations[2])
            # Check if the fifth element is "NEW"
            if len(combinations) > 4 and combinations[4] == "NEW":
                final[i].append(combinations[3] + "_" + combinations[4])
            else:
                final[i].append(combinations[3])
        else:
            # Append directly to the final list
            final[i].append(combinations[1])
            # Check if the fourth element is "NEW"
            if len(combinations) > 3 and combinations[3] == "NEW":
                final[i].append(combinations[2] + "_" + combinations[3])
            else:
                final[i].append(combinations[2])
    return final
```

```python
def get_programs(lst):
    programs = set()  # Usamos un set para evitar duplicados
    for item in lst:
        parts = item.split("_")
        # Agrupar cada par de elementos consecutivos
        for i in range(0, len(parts)-1, 2):  # Recorremos de 2 en 2
            program = f"{parts[i]}_{parts[i+1]}"  # Concatenamos cada par
            programs.add(program)  # Lo agregamos al set
    return programs
```

```python
def split_string(input_string):
    # Dividir la cadena por "_"
    parts = input_string.split("_")
    
    # Crear las dos partes deseadas
    part1 = "_".join(parts[:2])  # Los primeros dos elementos: 'VINA_OLD'
    part2 = "_".join(parts[2:])  # Los siguientes dos elementos: 'DOCK6_NEW'
    
    return {part1, part2}
```

```python
from itertools import combinations
def calc_ecr(*args, dir_excel):
    excel_temp = dir_excel.copy()
    ecr_values = [sum(values) for values in zip(*args)]

    col_name = "exp_" + "_".join([f"p{i+1}" for i in range(len(args))])
    excel_temp[col_name] = ecr_values
    df_temp = excel_temp.sort_values(by=[col_name],axis = 0, ascending=False)

    df_temp[f"ECR_{len(args)}p"] = np.arange(1, len(df_temp) +1)
    df_temp = df_temp.sort_values(by=['IX'], axis = 0)
    df_temp.drop(columns=[col_name])
    return pd.to_numeric(df_temp[f"ECR_{len(args)}p"].tolist())


def calc_ecrs(dir_excel_8p):
    excel_8p = dir_excel_8p.copy()
    excel_8p = excel_8p.sort_values(by='IX',axis = 0, ascending=True)

    
    programs = [
        ("pl_1", excel_8p['exp_PL_1'].tolist()),
        ("pl_2", excel_8p['exp_PL_2'].tolist()),
        ("ad_1", excel_8p['exp_AD_1'].tolist()),
        ("ad_2", excel_8p['exp_AD_2'].tolist()),
        ("dock6_1", excel_8p['exp_dock6_1'].tolist()),
        ("dock6_2", excel_8p['exp_dock6_2'].tolist()),
        ("vina_1", excel_8p['exp_vina_1'].tolist()),
        ("vina_2", excel_8p['exp_vina_2'].tolist()),
    ]
    n = len(programs)

    new_columns = {}

    for i in range(2, n+1):  # Tamaños de combinaciones (1 a 8)
        for selected_programs in combinations(programs, i):
            
            col_name = "ecr_" + "_".join([name for name, _ in selected_programs])
            
            new_columns[col_name] = calc_ecr(*[data for _, data in selected_programs], dir_excel=dir_excel_8p)
            
    excel_8p = excel_8p.sort_values(by='IX',axis = 0, ascending=True)
    excel_8p = excel_8p.assign(**new_columns)


    
    return excel_8p
```

```python
def rename_ecr_columns(df):
    new_columns = {}
    
    for col in df.columns:
        if col.startswith("ecr_"):
            parts = col.split("_")
            prefix = parts[:1]  # "ecr"
            grouped = [tuple(parts[i:i+2]) for i in range(1, len(parts), 2)]  # Agrupar cada 2 elementos
            sorted_grouped = sorted(grouped, key=lambda x: x[0])  # Ordenar por el primer elemento de cada par
            new_name = "_".join(prefix + [item for group in sorted_grouped for item in group])  # Aplanar
            new_columns[col] = new_name

    df = df.rename(columns=new_columns)
    return df
```

```python
def calc_ecr2(exp_p1,exp_p2,dir_excel_5p):

    excel_temp = (dir_excel_5p)
    
    zipped_lists_2 = zip(exp_p1, exp_p2)
    ecr_2 = [x1 + x2  for (x1, x2) in zipped_lists_2] # para q sume los de la misma linea
    excel_temp['exp_p1_p2'] = ecr_2
    
    # Agrego la columna del rank exp y vuelvo a ordenar por mol
    df_temp = excel_temp.sort_values(by =['exp_p1_p2'], axis=0, ascending=False)
    cons_rank = np.arange(1, len(df_temp)+1)
    df_temp['ECR_2p']=cons_rank
    df_temp = df_temp.sort_values(by=['IX'], axis=0)
    df_temp.drop(columns=['exp_p1_p2'])
    
    return pd.to_numeric(df_temp['ECR_2p'].tolist())
```
